```python
from max_subarray import cubic_algorithm, quadratic_algorithm_1, quadratic_algorithm_2, divide_and_conquer_algorithm, linear_algorithm
import numpy as np

randtests = []
for i in range(100):
    x = np.random.randint(-10,10,200)
    results = cubic_algorithm(x)
    if (results != quadratic_algorithm_1(x)):
        randtests += [x]
    if (results != quadratic_algorithm_2(x)):
        randtests += [x]
    if (results != divide_and_conquer_algorithm(x)):
        randtests += [x]
    if (results != linear_algorithm(x)):
        randtests += [x]
        
print(len(randtests))
```
