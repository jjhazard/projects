```python
import pandas as pd
import numpy as np
from D import DecisionTree, Node

D = pd.read_csv('xTrain.csv')
E = pd.read_csv('yTrain.csv')
D["y"] = E.values
DT = DecisionTree(D[:1000])
```

```python
def rules(rule, node, depth):
    if (node.name == "Leaf"):
        rule += [" then ", str(node.data[0])," with depth ", str(depth)]
        for i in rule:
            print(i,end="")
        print("")
    else:
        lrule = rule + ["&(",node.name,"<",str(node.data[0]),")"]
        rules(lrule, node.branches[0],depth+1)
        rrule = rule + ["&(",node.name,">",str(node.data[0]),")"]
        rules(rrule, node.branches[1],depth+1)
        
        
rules([],DT.head,1)
```

```python
def classify(D, tree):
        results = []
        for row in range(D.shape[0]):
            hold = tree.head
            x = D.iloc[row,:]
            while (hold.branches != []):
                hold = hold.compare(x[hold.name])
            results += [hold.data[0]]
        return results
x = D
guesses = classify(x, DT)
results = x["y"].tolist()
correct = 0
confMat = {'Guess1': [0,0,0,0,0], 'Guess2': [0,0,0,0,0], 'Guess3': [0,0,0,0,0], 'Guess4': [0,0,0,0,0], 'Guess5': [0,0,0,0,0]}
confMat = pd.DataFrame(confMat)
for i in range(len(guesses)):
    if (results[i] == guesses[i]):
        correct += 1
    confMat.iloc[(results[i]-1),(guesses[i]-1)] += 1
print(correct, correct/len(guesses))


print(confMat)
```
